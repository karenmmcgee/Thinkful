```python
#find the unique value of Pclass
print(df.Pclass.unique())
```

```python
#create a new data frame called features and fill it with Pclass
features = pd.get_dummies(df['Pclass'])
```

```python
#add age to the features column
features['Age']=df.Age
```

```python
#add gender to the features column
features['Gender']=df.Sex 
```

```python
#add Parent child to the featurs column
features['Parent Child']=df.Parch>0  
```

```python
#add fare to the features column
features['Fare']=df.Fare
```

```python
#add cabin to the features column
features['Cabin'] = df.Cabin
```
