```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('D5 SHSAT Registrations and Testers.csv')
print(df)
```

```python
plt.plot(df['Enrollment on 10/31'], color='purple')
plt.plot(df['Number of students who registered for the SHSAT'], color='green')
plt.plot(df['Number of students who took the SHSAT'], color='blue')
plt.ylabel('Values')
plt.title('Enrollment as of 10/31')
plt.show()
```

```python
plt.scatter(x=df['Enrollment on 10/31'], y=df['Number of students who registered for the SHSAT'])
plt.show()
```

```python
plt.scatter(
    x=df['Enrollment on 10/31'],
    y=df['Number of students who registered for the SHSAT'],
    color='purple',
    marker='x', s=10
)
plt.scatter(
    x=df['Enrollment on 10/31'],
    y=df['Number of students who took the SHSAT'],
    color='green',
    marker='x', s=10
)
plt.show()
```

```python
df.plot(kind='scatter', x='Number of students who registered for the SHSAT',y='Number of students who took the SHSAT')
df.plot(kind='line')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Number of students who registered for the SHSAT'], color='purple')
plt.ylabel('Values')
plt.title('Number of students who registered for the SHSAT')

plt.subplot(1, 2, 2)
plt.plot(df['Number of students who took the SHSAT'], color='green')
plt.ylabel('Values')
plt.title('Number of students who took the SHSAT')
plt.show()
```
