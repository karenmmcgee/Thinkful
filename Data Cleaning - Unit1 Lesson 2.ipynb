```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the file into a data frame
# use the following format when reading the file
#read_csv('file', encoding = "ISO-8859-1"), or alternatively encoding = "utf-8"
#df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv',encoding = 'ISO-8859-1')
#This will automatically consider the empty fields as empty strings ' '

df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv',encoding = 'ISO-8859-1' , na_filter=False)
df.head()
```

```python
#Checking for missing values
df.isnull().head(20)
```

```python
print(df.apply(lambda x:x.fillna(" ")))
```

```python
#Remove the \n and \r from the PMID/PMCID column
print(df.apply(lambda x: str(x).strip('\\n')))
```

```python
# Drop only rows where all values are missing.
print(df.dropna(how='all'))
```

```python
# Drop all rows that have any missing values in the 'gender' or 'height' columns.
print(df.dropna(subset=['PMID/PMCID']))
```
