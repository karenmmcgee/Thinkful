```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
df = pd.read_csv('speed-camera-violations.csv')
```

```python
#Lets describe the data to see what the statistical data set contains
df.describe()
```

```python
df.head(10)
```

```python
#looking at the data, I want to make the Violation data a true continuous variable
#therefore, I will use the next few cells to remove the time componet from the cell block
df['VIOLATION DATE'][0]
```

```python
#split the cell for each -
df['VIOLATION DATE'][0].split('-')
```

```python
#test that I split the cell correctly
df['VIOLATION DATE'][0].split('-')[0:2]
```

```python
#Now rejoin the split cell to contain just the month and year
'-'.join(df['VIOLATION DATE'][0].split('-')[0:2])
```

```python
#create a function to perform the split operation on all of the records for the Violation date column
def new_string(String1):
    return '-'.join(String1.split('-')[0:2])
```

```python
#Add a column to the table to reflect the changes of the split
df['VIOLATION_MONTH'] = df['VIOLATION DATE'].apply(lambda x: new_string(x))
```

```python
#verify the new column was added 
df
```

```python
#Graph violations by address and violation month
trace = go.Scattergl(
    x = df.groupby(['ADDRESS','VIOLATION_MONTH'])['VIOLATIONS'].sum(),
    y = df['VIOLATION_MONTH'].unique(),
    mode = 'markers',
    marker = dict(
        color = '#FFBAD2',
        line = dict(width = 1)
    )
)
data = [trace]

layout= go.Layout(
    title= 'Total violations by month',
    xaxis= dict(
        title= 'Violations',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Violations by Month',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False
)
fig= go.Figure(data=data, layout=layout)
py.iplot(fig)
#py.iplot(data, filename='compare_webgl')
```

```python
# graph all three variables Total violations, violation  month and address
trace = go.Heatmap(z=df.groupby(['ADDRESS','VIOLATION_MONTH'])['VIOLATIONS'].sum(),
                   y=df['VIOLATION_MONTH'].unique(),
                   x=df['ADDRESS'].unique())
data=[trace]

layout= go.Layout(
    title= 'Violations by month and Address',
    xaxis= dict(
        title= 'Violations by location',
        ticklen= 5,
        zeroline= True,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Violations by Month',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= True,
    margin= dict(l=80,r=80)
)
fig= go.Figure(data=data, layout=layout)
py.iplot(fig)
#py.iplot(data, filename='labelled-heatmap')
```
