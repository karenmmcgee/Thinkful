```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#Load the file into a data frame
df = pd.read_csv('BlackFriday.csv')
df.head()
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

ax = sns.boxplot(x='Age',y='Purchase',hue='Gender',data=df,palette='pastel')  

plt.title('Plot using seaborn: Purchase amount by Gender')

sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Purchase amount')
plt.legend()
plt.show()
```

```python
#Pros of graph: Graph looks great and shows the distribution of purchase amount and age
#Cons of graph: The legend is hidden behind the data. 
```

```python
# Setting the overall aesthetic.
sns.set(style="darkgrid")

g = sns.factorplot(x="Age", y="Purchase", hue="Gender", data=df,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Purchase amount")
g.set_xlabels("")
plt.title('Barplot using seaborn: Purchase amount by Gender')
plt.legend()
plt.show()
```

```python
#Pros of graph: Graph looks great and shows the distribution of purchase amount by age and gender 
    #and the legend sits outside of the graph for easy view. 
#Cons of graph: graph could be scaled to bigger ratio.
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.factorplot(x="Age", y="Purchase", hue="Gender", data=df,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Purchase Amount")
g.set_xlabels("")
plt.title('Pointplot using seaborn: Purchase amount by Gender')
plt.show()
```

```python
#Pros of graph: I like this graph the best as it shows the distribution of purchase amount by age and gender 
    #in a much more intuitve and visual way. The legend sits outside of the graph for easy view. 
#Cons of graph: None
```

```python
df2 = pd.read_csv('avocado.csv')
df2.head()
```

```python
# Setting the overall aesthetic for 2 variables.

g = sns.lmplot(y='AveragePrice', # Variable 1.
               x='year', # Variable 2.
               data=df2, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Average price of Avocados")
g.set_xlabels("Over Time")
plt.title('Scatterplot: Average price of avocados over time')
plt.show()

g = sns.lmplot(y='AveragePrice', 
               x='year',
               data=df2, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("Average price of Avocados")
g.set_xlabels("Over Time")
plt.title('Scatterplot with regression line: Average price of avocados over time')
plt.show()
```

```python
#Pros of graph: Graph looks visually good.
#Cons of graph: the data dosn't allow the graph to shows the distribution of purchase amount by age and gender 
    #in an easy to see scatter plot. 
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.factorplot(x="AveragePrice", y="region", data=df2,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Region")
g.set_xlabels("")
plt.title('Pointplot using seaborn: Purchase price of avocado by region')
plt.show()
```

```python
#Pros of graph: Graph looks visually good and overall you get an idea of how the data aligns
#Cons of graph: the region data is too much for the graph and makes the data looks squeezed. 
```
