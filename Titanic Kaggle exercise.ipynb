```python
import pandas as pd
import numpy as np
```

```python
#load the data into a data frame
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
#make Survived our target column
columns_target=['Survived']
```

```python
#set the following columns for our training set
columns_train=['Age','Pclass','Sex','Fare']
```

```python
#set x = to our training set
X=data[columns_train]
```

```python
#set y = to our column target
Y=data[columns_target]
```

```python
#data preprocessing stage
#check the training columns for null values
#if we find them replace there values with the median 
X['Sex'].isnull().sum()
```

```python
X['Pclass'].isnull().sum()
```

```python
X['Fare'].isnull().sum()
```

```python
X['Age'].isnull().sum()
```

```python
#We found null values for the Age training column
X['Age']=X['Age'].fillna(X['Age'].median())
X['Age'].isnull().sum()
```

```python
#set the Sex training column to 0 or 1 as the linear model can't take in a string
d={'male':0, 'female':1}
```

```python
X['Sex']=X['Sex'].apply(lambda x:d[x])
```

```python
#verify the training column Sex = 0 or 1
X['Sex'].head()
```

```python
# print the data frame to see our changes
X
```

```python
X.head()
```

```python
#now train our data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
#now fit and print our data
from sklearn import svm
clf=svm.LinearSVC()
clf.fit(X_train, Y_train)
print (clf)
```

```python
#predict our test data
print (clf.predict(X_test[0:1]))
```

```python
print (clf.predict(X_test[0:10]))
```

```python
#verify our prediction 
print (clf.score(X_test,Y_test))
```
