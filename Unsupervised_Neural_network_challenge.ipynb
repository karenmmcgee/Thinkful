```python
#load the data into a dataframe.
df = pd.read_csv("amazon_cells_labelled.txt", delimiter= '\t', header=None)
df.head()
```

```python
#Add column names
df.columns = ['message', 'sentiment']
```

```python
#verify the columns have been created. 
df.head()
```

```python
#for i in range(0,len(df)):
 #   df.loc[0:i,['message']]
  #  print(df.loc[i,['message']])
```

```python
#use keywords to build predictive variables
keywords = ['Good', 'Great', 'Poor', 'Not', 'No', 'Best', 'Excellent', 'Disappoint','Waste','Problem']

#keywords = tf1.words


for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matchin
    df[str(key)] = df.message.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
```

```python
#look for upper case words
df['allcaps'] = df.message.str.isupper()
```

```python
#make sure message dosn't have any bad data
df['message'] = (df['message'] == 'message')
```

```python
#verify dataframe before modeling it again
df[keywords].head()
```

```python
df.head()
```

```python
#create the X and y values
X = df[keywords + ['allcaps']]
y = df['sentiment']
```

```python
# Alright! We've done our prep, let's build the model.
# Neural networks are hugely computationally intensive.
# This may take several minutes to run.

# Import the model.
from sklearn.neural_network import MLPClassifier

# Establish and fit the model, with a single, 1000 perceptron layer.
mlp = MLPClassifier(hidden_layer_sizes=(1000,))
mlp.fit(X, y)
```

```python
print('Overal R^2 score: {}'.format(mlp.score(X, y)))
```
